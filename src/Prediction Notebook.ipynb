```python
matches["hero_bans_code"] = matches["hero_bans"].astype("category").cat.codes
matches["phase_code"] = matches["phase"].astype("category").cat.codes
matches["team_code"] = matches["team"].astype("category").cat.codes
matches["player_code"] = matches["player"].astype("category").cat.codes
matches["role_code"] = matches["Role"].astype("category").cat.codes
#converting some of the strings to codes like the hero code and each hero has it's number id same goes for the others
```

```python
rf = RandomForestClassifier(n_estimators=610, min_samples_split=98, random_state=1)
#random forest is just a number of decision trees that each tree has a slightly different parameters
#n_estimators = the number of decision trees we wanna make (the higher the number, the higher potential accuracy it'll be) but i'll take longer to load
#min_samples_split is the number of leafs we have in the tree before we split the node (the higher the number, the less likely we will overfit) 
#^ but the lower the accuracy our training data will be
```

```python
train = matches[matches["match_date"] < '2025-07-30'] 
#spliting the data (training data) before Stage 3
```

```python
test = matches[matches["match_date"] > '2025-07-30']
#spliting the data (test data) Stage 3
```

```python
predictors = [
    "Damage Dealt", "Time Played", "Damage Mitigated", "K/D Ratio",
    "Healing Done", "Deaths", "Final Blows", "Eliminations",
    "phase_code", "team_code", "player_code", "role_code", "hero_bans_code"
]
#making predictors
```

```python
rf.fit(train[predictors], train["Result"])
#training a random forest using our predictors trying to predict our result which is "Win(1)" or "Lose(0)"
```

```python
probs = rf.predict_proba(test[predictors])[:, 1]
test_with_preds = test.copy()
test_with_preds["proba"] = probs
#adding win probabilities into the test set
```

```python
team_level = (
    test_with_preds.groupby(["match_id", "team"])
    .agg({
        "Result": "first",         
        "proba": "mean"            
    })
    .reset_index()
)
team_level["prediction"] = (team_level["proba"] >= 0.5).astype(int)
acc_team = accuracy_score(team_level["Result"], team_level["prediction"])
print("Team-level accuracy:", acc_team)

print(pd.crosstab(team_level["Result"], team_level["prediction"]))
#checking accuracy at team level
#giving us a 92% accuracy 
```

```python
match_level = (
    test_with_preds.groupby("match_id")
    .apply(lambda g: pd.Series({
        "actual": g.loc[g["Result"] == 1, "team"].values[0],
        "predicted": g.loc[g["proba"].idxmax(), "team"]
    }))
    .reset_index()
)

match_acc = accuracy_score(match_level["actual"], match_level["predicted"])
print("Match-level accuracy:", match_acc)
print(pd.crosstab(match_level["actual"], match_level["predicted"]))
#collapsing to match level so we only have one prediction per match
#the crosstab is confusing so run the next code to see how many we predicted wrong
#96% accuracy on a match level pretty good.
```

```python
match_level["correct"] = match_level["actual"] == match_level["predicted"]

print(pd.crosstab(match_level["correct"], columns="count"))
#how many matches predicted right vs wrong
#26 out of 27! basically we can try to predict any match in the future
```

```python
def predict_upcoming_matches(model, matches_data, predictors, upcoming_pairs):
    results = []
    for team1, team2 in upcoming_pairs:
        team1_data = matches_data[matches_data["team"] == team1].sort_values("match_date").iloc[-1]
        team2_data = matches_data[matches_data["team"] == team2].sort_values("match_date").iloc[-1]
        future_match = pd.DataFrame([
            team1_data[predictors],
            team2_data[predictors]
        ])
        probs = model.predict_proba(future_match)[:, 1]
        results.append({
            "match": f"{team1} vs {team2}",
            "team1": team1,
            "team1_proba": probs[0],
            "team2": team2,
            "team2_proba": probs[1],
            "predicted_winner": team1 if probs[0] > probs[1] else team2
        })
    
    return pd.DataFrame(results)
# function to predict upcoming matches using latest stats
```

```python
cols = [
    "Damage Dealt", "Time Played", "Damage Mitigated", "K/D Ratio",
    "Healing Done", "Deaths", "Final Blows", "Eliminations",
    "Solo Kills/Deaths Ratio", "Damage Mitigated/Eliminations",
    "Damage as percent of total output", "Healing Done/Eliminations",
    "Damage Mitigated/Deaths", "Damage Mitigated as percent of total output",
    "Objective Time"
]

new_cols = [col + "_rolling" for col in cols]

def rolling_average(group, cols, new_cols):
    group = group.sort_values("match_date")
    rolling = group[cols].rolling(5, min_periods=1).mean()
    group[new_cols] = rolling
    return group

matches_rolling = (
    matches.groupby(["region", "team"], group_keys=False)
    .apply(lambda x: rolling_average(x, cols, new_cols))
    .reset_index(drop=True)
)
#building rolling averages for teams across last 5 matches
#rolling averages are now built separately for each team inside its region
```

```python
def predict_upcoming_matches_region(model, matches_data, predictors, upcoming_pairs, region=None):
    results = []
    for team1, team2 in upcoming_pairs:
        team1_hist = matches_data[(matches_data["team"] == team1)]
        team2_hist = matches_data[(matches_data["team"] == team2)]
        if region is not None:
            team1_hist = team1_hist[team1_hist["region"] == region]
            team2_hist = team2_hist[team2_hist["region"] == region]

        if team1_hist.empty or team2_hist.empty:
            results.append({
                "match": f"{team1} vs {team2}",
                "team1": team1,
                "team1_proba": None,
                "team2": team2,
                "team2_proba": None,
                "predicted_winner": "MISSING DATA"
            })
            continue

        team1_data = team1_hist.sort_values("match_date").iloc[-1]
        team2_data = team2_hist.sort_values("match_date").iloc[-1]

        future_match = pd.DataFrame([
            team1_data[predictors],
            team2_data[predictors]
        ])

        probs = model.predict_proba(future_match)[:, 1]

        results.append({
            "match": f"{team1} vs {team2}",
            "team1": team1,
            "team1_proba": probs[0],
            "team2": team2,
            "team2_proba": probs[1],
            "predicted_winner": team1 if probs[0] > probs[1] else team2
        })
    return pd.DataFrame(results)
#function to predict upcoming matches by checking last match data for each team (with region filter if given)
```

```python
def predict_upcoming_matches_region(model, matches_data, predictors, upcoming_pairs, region=None):
    results = []
    for team1, team2 in upcoming_pairs:
        # filter teams (and region if specified)
        team1_hist = matches_data[matches_data["team"] == team1]
        team2_hist = matches_data[matches_data["team"] == team2]
        if region is not None:
            team1_hist = team1_hist[team1_hist["region"] == region]
            team2_hist = team2_hist[team2_hist["region"] == region]

        if team1_hist.empty or team2_hist.empty:
            results.append({
                "match": f"{team1} vs {team2}",
                "team1": team1,
                "team1_proba": None,
                "team2": team2,
                "team2_proba": None,
                "predicted_winner": "MISSING DATA"
            })
            continue

        team1_data = team1_hist.sort_values("match_date").iloc[-1]
        team2_data = team2_hist.sort_values("match_date").iloc[-1]

        future_match = pd.DataFrame([
            team1_data[predictors],
            team2_data[predictors]
        ])
        probs = model.predict_proba(future_match)[:, 1]

        total = probs[0] + probs[1]
        team1_norm = probs[0] / total
        team2_norm = probs[1] / total

        results.append({
            "match": f"{team1} vs {team2}",
            "team1": team1,
            "team1_proba": team1_norm,
            "team2": team2,
            "team2_proba": team2_norm,
            "predicted_winner": team1 if team1_norm > team2_norm else team2
        })
    return pd.DataFrame(results)
#function to predict upcoming matches using latest match data with normalized probabilities (optionally filtered by region)
```
